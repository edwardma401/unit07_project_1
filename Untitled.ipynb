```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_one = "Resources/covid_19_data.csv"
file_one_df = pd.read_csv(file_one, encoding = "ISO-8859-1")
file_sorted = file_one_df.sort_values(['Confirmed'], ascending=False)
file_sorted
```

```python
loced = file_sorted.loc[file_sorted['Confirmed'] != 0 , :]
```

```python
loced
```

```python
merged_rows = loced.groupby(['Country/Region']).agg({"Confirmed": "sum"},
                                                   {"Deaths": "sum"},
                                                   {"Recovered" : "sum"})
```

```python
merged_rows
```

```python
merged_rows.plot.box()
```
