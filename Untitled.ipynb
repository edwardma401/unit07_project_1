```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_one = "Resources/covid_19_data.csv"
file_one_df = pd.read_csv(file_one, encoding = "ISO-8859-1")
file_sorted = file_one_df.sort_values(['Confirmed'], ascending=False)
file_sorted
```

```python
no_dup = file_sorted.drop_duplicates(subset='Province/State', keep='first')
no_dup # This process should remove the duplicates of Province/State, Since all we need is the data of the one Last Update.
```

```python
loc_no_dup = no_dup.loc[no_dup['Country/Region'] != "Others" , :]
```

```python
merged_rows = loc_no_dup.groupby(['Country/Region']).agg({"Confirmed": "sum",
                                                   "Deaths": "sum",
                                                   "Recovered" : "sum"})
```

```python
merged_rows # Shall we drop 'Others' in Country/Region? 
# Dropped 'Others'
```

```python
list_confirmed = merged_rows['Confirmed'].tolist()
print(type(list_confirmed))
# print(list_confirmed)
# print(type(merged_rows))
x_axis = merged_rows.index.tolist()
print(x_axis)
print(len(x_axis))
```

```python
plt.plot(x_axis, list_confirmed)
plt.xticks(rotation='vertical')
plt.ylabel('Confirmed Cases')
plt.title("Counts of Confirmed Covid-19 cases in Countries")
plt.show()
```

```python
test_file = 'Resources/owid-covid-data.csv'
test_file_df = pd.read_csv(test_file)
test_file_df.head()
```

```python
test_file_df_del = test_file_df[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'cvd_death_rate', 'median_age', 'total_tests', 'gdp_per_capita', 'population', 'population_density', 'life_expectancy']]
test_file_df_del_no_nan = test_file_df_del.dropna(how='any')
```

```python
test_file_df_no_dup = test_file_df_del_no_nan.drop_duplicates(subset='location', keep='last')
test_file_df_no_dup_loc = test_file_df_no_dup.loc[(test_file_df_no_dup['location'] != "World") & (test_file_df_no_dup['location'] != "International"), :]
test_file_df_no_dup_loc
```

```python
test_file_df_no_dup_loc_1 = test_file_df_no_dup_loc.loc[test_file_df_no_dup_loc['date'] != "2020-04-29", :]
test_file_df_no_dup_loc_1
```

```python
sorted_df = test_file_df_no_dup_loc_1.sort_values("total_tests", ascending=False)
sorted_df
```

```python
sorted_df_1 = sorted_df.iloc[:10, :]
sorted_df_1
```

```python
list_tests = sorted_df_1['total_tests'].tolist()
print(len(list_tests))

x_axis = sorted_df_1['location'].tolist()
print(x_axis)
print(len(x_axis))
```

```python
plt.bar(x_axis, list_tests)
plt.tight_layout
plt.xticks(rotation=45)
plt.title('Top 10 tests reported')
plt.ylabel('Number of tests (10 million)')
plt.show()
# Top Ten
```

```python
list_cases = sorted_df_1['total_cases'].tolist()
print(len(list_tests))
```
