```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_one = "Resources/covid_19_data.csv"
file_one_df = pd.read_csv(file_one, encoding = "ISO-8859-1")
file_sorted = file_one_df.sort_values(['Confirmed'], ascending=False)
file_sorted.head()
```

```python
no_dup = file_sorted.drop_duplicates(subset='Province/State', keep='first')
no_dup.head() # This process should remove the duplicates of Province/State, Since all we need is the data of the one Last Update.
```

```python
loc_no_dup = no_dup.loc[no_dup['Country/Region'] != "Others" , :]
```

```python
merged_rows = loc_no_dup.groupby(['Country/Region']).agg({"Confirmed": "sum",
                                                   "Deaths": "sum",
                                                   "Recovered" : "sum"})
merged_rows['Deaths_Rate'] = round(merged_rows['Deaths']/merged_rows['Confirmed'],3)
```

```python
merged_rows.head() # Shall we drop 'Others' in Country/Region? 
# Dropped 'Others'
```

```python
test_file = 'Resources/owid-covid-data.csv'
test_file_df = pd.read_csv(test_file)
test_file_df.head()
```

```python
ddd = test_file_df[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'cvd_death_rate', 'median_age', 'total_tests', 'gdp_per_capita', 'population', 'population_density', 'life_expectancy']]
ddd.head()
```

```python
no_nan = ddd.dropna(how='any')
```

```python
no_dup = no_nan.drop_duplicates(subset='location', keep='last')
no_dup_loc = no_dup.loc[(no_dup['location'] != "World") & (no_dup['location'] != "International"), :]
no_dup_loc.head()
```

```python
no_dup_loc_1 = no_dup_loc.loc[no_dup_loc['date'] != "2020-04-29", :]
no_dup_loc_1.head()
```

```python
sorted_df = no_dup_loc_1.sort_values("total_tests", ascending=False)
sorted_df.head()
```

```python
sorted_df_1 = sorted_df.iloc[:20, :]
sorted_df_1.head()
```

```python
list_tests = sorted_df_1['total_tests'].tolist()
print(len(list_tests))

x_axis = sorted_df_1['location'].tolist()
print(x_axis)
print(len(x_axis))
```

```python
list_tests_2 = [round(x/1000000,1) for x in list_tests] ## zoom out 1 million
plt.bar(x_axis, list_tests_2) ## draw a bar 
plt.tight_layout
plt.xticks(rotation='vertical')
plt.yticks(range(0, 40,5)) ## set the y ticks 
plt.title('Top 10 tests reported')
plt.ylabel('Number of tests (1 million)') ## set y label and unit
for x,y in zip(x_axis, list_tests_2):
    ## draw data labels
    plt.text(x, y+0.3, str(y), ha='center', va='bottom', fontsize=8)
plt.show()
# Top Ten
```

```python
list_cases = sorted_df_1['total_cases'].tolist()
print(len(list_tests))
```

```python
list_cases_2 = [round(x/1000000,1) for x in list_cases] ## zoom out 1 million
plt.bar(x_axis, list_cases_2) ## draw bar 
plt.tight_layout
plt.xticks(rotation='vertical')
plt.yticks(np.arange(0, 3.5, 0.5)) ## set y ticks
plt.title('Top 10 cases reported')
plt.ylabel('Number of cases (million)')
for x,y in zip(x_axis, list_cases_2):
    ## draw data labels
    plt.text(x, y+0.1, str(y), ha='center', va='bottom', fontsize=8)
plt.show()
```

```python
df_death_10 = merged_rows.sort_values(by = 'Deaths_Rate', ascending = False)
df_death_10 = df_death_10.head(10)
country_list = list(df_death_10.index)
death_rate_list = list(df_death_10['Deaths_Rate'])
plt.bar(country_list, death_rate_list)
plt.tight_layout
plt.xticks(rotation = 'vertical')
plt.yticks(np.arange(0, 0.18, 0.02))
plt.title('Top 10 Deaths_Rate reported')
plt.ylabel('Deaths Rate')
for x,y in zip(country_list, death_rate_list):
    plt.text(x, y, str(round(y*100,1)) + '%', ha='center', va='bottom', fontsize=8)
plt.show()

```
