```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_one = "Resources/covid_19_data.csv"
file_one_df = pd.read_csv(file_one, encoding = "ISO-8859-1")
file_sorted = file_one_df.sort_values(['Confirmed'], ascending=False)
file_sorted
```

```python
no_dup = file_sorted.drop_duplicates(subset='Province/State', keep='first')
no_dup # This process should remove the duplicates of Province/State, Since all we need is the data of the one Last Update.
```

```python
merged_rows = no_dup.groupby(['Country/Region']).agg({"Confirmed": "sum",
                                                   "Deaths": "sum",
                                                   "Recovered" : "sum"})
```

```python
merged_rows # Shall we drop 'Others' in Country/Region?
```

```python
list_confirmed = merged_rows['Confirmed'].tolist()
print(type(list_confirmed))
# print(list_confirmed)
# print(type(merged_rows))
x_axis = merged_rows.index.tolist()
print(x_axis)
```

```python
plt.plot(x_axis, list_confirmed)
plt.xticks(rotation='vertical')
plt.ylabel('Confirmed Cases')
plt.title("Counts of Confirmed Covid-19 cases in Countries")
plt.show()
```

```python
list_death = merged_rows['Deaths'].tolist()
# print(type(list_death))
# print(list_death)
plt.plot(x_axis, list_death)
plt.xticks(rotation='vertical')
plt.title("Counts of Deaths due to Covid-19 in Countries")
plt.ylabel("Number of Deaths")
plt.show()
```

```python
list_recovered = merged_rows['Recovered'].tolist()
# print(list_recovered)
plt.plot(x_axis, list_recovered)
plt.xticks(rotation='vertical')
plt.title('Counts of Recovered from Covid-19 in Countries')
plt.ylabel('Recovered Cases')
plt.show()
```

```python
print(merged_rows['Confirmed'].describe())
plt.boxplot(merged_rows['Confirmed'])
plt.ylabel('Confirmed Cases')
plt.title('Possible outliers of Confirmed Cases in Countries')
# plt.ylim(0, 500000)
plt.show()
print("The median of this bar graph is : {0}".format(merged_rows['Confirmed'].median() ))
```
